```python
# Import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
office_ep = pd.read_csv('datasets/office_episodes.csv')

# Print DataFrame
print(office_ep.head())

# DataFrame Info
print(office_ep.info())
```

```python
fig = plt.figure()
# Create scatter plot
plt.style.use('seaborn-whitegrid')
plt.scatter(office_ep.episode_number, office_ep.viewership_mil)

#Create title
plt.title("The Office Viewership per Episode")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```

```python
# Create Color System
# Ratings < 0.25 are colored "red"
# Ratings >= 0.25 and < 0.50 are colored "orange"
# Ratings >= 0.50 and < 0.75 are colored "lightgreen"
# Ratings >= 0.75 are colored "darkgreen"

# Define colors list
colors = []

# Iterate over rows of office
for lab, row in office_ep.iterrows() :
    if row['scaled_ratings'] < 0.25 :
        colors.append("red")
    elif 0.25 <=  row['scaled_ratings']  < 0.50 :
        colors.append("orange")
    elif 0.50 <=  row['scaled_ratings']  < 0.75 :
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
        
# Inspect the first 10 values in colors list        
colors[0:10]
```

```python
# Create Sizing System
# episodes with guest appearances have a marker size of 250
# episodes without are sized 25

# Define Sizes list
sizes = []

# Iterate over rows of office
for lab, row in office_ep.iterrows() :
    if row['has_guests'] == True :
        sizes.append(250)
    else: 
        sizes.append(25)
        
# Inspect the first 10 values in sizes list
sizes[0:10]
```

```python
# Create Shape System
# episodes with guest appearances have a star shape

# Define Shape list
shapes = []

# Iterate over rows of office
for lab, row in office_ep.iterrows() :
    if row['has_guests'] == True :
        shapes.append("*")
    
```

```python
fig = plt.figure()

# Create Scatter Plot
plt.style.use('seaborn-whitegrid')
plt.scatter(office_ep.episode_number, office_ep.viewership_mil,marker="*",c=colors,s=sizes)
plt.rcParams['figure.figsize'] = [11, 7]

# Create Title
plt.title("Popularity, Quality, and Guest Appearances on the Office")

# Create Labels
plt.xlabel("Episode Number")
plt.ylabel ("Viewership (Millions)")

# Show Plot
plt.show()
```
